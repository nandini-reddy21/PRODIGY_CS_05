```python
from scapy.all import sniff, IP, TCP, UDP
from scapy.layers.http import HTTPRequest
import time
capture_running = True

def stop_sniffing(packet):
    global capture_running
    return not capture_running
```

```python
def process_packet(packet):
  if IP in packet:
        ip_layer = packet[IP]
        print(f"\n[IP Packet] Source: {ip_layer.src}, Destination: {ip_layer.dst}")
  if TCP in packet:
        tcp_layer = packet[TCP]
        print(f"[TCP Packet] Source Port: {tcp_layer.sport}, Destination Port: {tcp_layer.dport}")
  if UDP in packet:
        udp_layer = packet[UDP]
        print(f"[UDP Packet] Source Port: {udp_layer.sport}, Destination Port: {udp_layer.dport}")
  if packet.haslayer(HTTPRequest):
        http_layer = packet[HTTPRequest]
        print(f"[HTTP Request] Host: {http_layer.Host.decode()}, Path: {http_layer.Path.decode()}")
```

```python
print("Starting packet capture for 10 seconds...")
start_time = time.time()
while capture_running:
    sniff(prn=process_packet, store=False, stop_filter=stop_sniffing, timeout=1)
    if time.time() - start_time > 10:
        capture_running = False
```
